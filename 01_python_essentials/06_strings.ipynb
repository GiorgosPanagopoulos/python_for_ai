# 6. Strings

Python strings are **immutable** (like Java), but Python has no `char` type — single characters are just strings of length 1. Python also supports powerful string formatting with f-strings and provides many built-in methods.

This notebook covers: common string methods, split/join, membership, formatting, multi-line strings, slicing, and character checks.

### 6.1 String Basics — Immutability

**☕ JAVA:** `String` is immutable. Use `StringBuilder` for mutation.
```java
String s = "Hello";
s = s + " World";  // Creates a new String object
char c = s.charAt(0);
```

**🐍 PYTHON:** Strings are also immutable. No `char` type — use indexing.
```python
s = "Hello"
s = s + " World"    # Creates a new string object
c = s[0]            # 'H' (still a string, not char)
```

````python
s = "Hello"
print(f"First char: {s[0]}")         # 'H'
print(f"Last char:  {s[-1]}")        # 'o'
print(f"Type: {type(s[0])}")         # <class 'str'> — not char!

# Strings are immutable — this would fail:
# s[0] = 'h'  # TypeError: 'str' object does not support item assignment
````

### 6.2 Common Methods

**☕ JAVA:** `toUpperCase()`, `toLowerCase()`, `trim()`, `replace()`

**🐍 PYTHON:** Similar methods, slightly different names.

````python
text = "  Hello, World!  "

print(f"Upper:   {text.upper()}")                      # Java: toUpperCase()
print(f"Lower:   {text.lower()}")                      # Java: toLowerCase()
print(f"Strip:   '{text.strip()}'")                    # Java: trim()
print(f"Replace: {text.replace('World', 'Python')}")   # Same as Java
````

### 6.3 Split and Join

**☕ JAVA:**
```java
String[] parts = "a,b,c".split(",");
String joined = String.join("-", parts);
```

**🐍 PYTHON:** `split()` is the same, but `join()` is called **on the separator** — this is a common gotcha!
```python
parts = "a,b,c".split(",")
joined = "-".join(parts)     # Note: separator.join(list)
```

````python
# Split — same as Java
csv = "Alice,Bob,Charlie"
names = csv.split(",")
print(f"Split: {names}")              # ['Alice', 'Bob', 'Charlie']

# Split with no argument — splits on any whitespace
sentence = "  Hello   World  "
words = sentence.split()
print(f"Split whitespace: {words}")   # ['Hello', 'World']

# Join — called on the separator!
print(f"Join with '-': {'-'.join(names)}")     # Alice-Bob-Charlie
print(f"Join with ' ': {' '.join(words)}")     # Hello World
````

### 6.4 String Membership (`in`)

**☕ JAVA:** `str.contains("text")` / `str.indexOf("text") >= 0`

**🐍 PYTHON:** No `.contains()` method — use the `in` operator instead.

````python
text = "Hello, World!"

print(f"'World' in text: {'World' in text}")       # True
print(f"'Python' in text: {'Python' in text}")     # False

# startswith / endswith — same as Java
print(f"\nStarts with 'Hello': {text.startswith('Hello')}")   # True
print(f"Ends with '!': {text.endswith('!')}")                 # True

# find() — returns index or -1 (like Java indexOf)
print(f"\nIndex of 'World': {text.find('World')}")   # 7
print(f"Index of 'xyz':   {text.find('xyz')}")       # -1
````

### 6.5 String Formatting

**☕ JAVA:** `String.format("Hello %s, age %d", name, age)` or `"Hello " + name`

**🐍 PYTHON:** Three ways (f-strings are the modern standard):
```python
f"Hello {name}"              # f-string (Python 3.6+) ✅ preferred
"Hello {}".format(name)      # .format() method
"Hello %s" % name            # % formatting (old style)
```

````python
name = "Alice"
age = 30
price = 19.99

# f-strings — the modern way (recommended)
print(f"{name} is {age} years old")

# Expressions inside f-strings
print(f"Next year: {age + 1}")
print(f"Name upper: {name.upper()}")

# Formatting numbers
print(f"\nPrice: ${price:.2f}")       # 2 decimal places
print(f"Big number: {1000000:,}")     # Thousands separator
print(f"Percentage: {0.856:.1%}")     # As percentage
````

### 6.6 Multi-line Strings

**☕ JAVA 15+:** Text blocks with `"""..."""`
```java
String text = """
              Hello,
              World!
              """;
```

**🐍 PYTHON:** Triple-quoted strings — available since Python 1!
```python
text = """Hello,
World!"""
```

````python
multiline = """This is a
multi-line
string."""
print(multiline)

# Also useful for long messages
email = """
Dear Alice,

Thank you for your order.
Your reference number is 12345.

Best regards,
Bob
""".strip()
print(f"\n{email}")
````

### 6.7 String Slicing

Strings support the same slicing syntax as lists — `[start:stop:step]`.

**☕ JAVA:** `str.substring(0, 5)` — limited.

**🐍 PYTHON:** `str[0:5]` — plus negative indexing, step, and reverse.

````python
text = "Hello, World!"

print(f"First 5:  {text[:5]}")          # Hello
print(f"Last 6:   {text[-6:]}")          # orld!
print(f"Substring: {text[7:12]}")        # World
print(f"Every 2nd: {text[::2]}")         # Hlo ol!
print(f"Reversed:  {text[::-1]}")        # !dlroW ,olleH
````

### 6.8 String Checks

**☕ JAVA:** `Character.isDigit()`, `Character.isLetter()` — work per-character.

**🐍 PYTHON:** Built-in methods that check the **entire string**.

````python
print(f"'123'.isdigit():    {'123'.isdigit()}")       # True
print(f"'abc'.isalpha():    {'abc'.isalpha()}")       # True
print(f"'abc123'.isalnum(): {'abc123'.isalnum()}")    # True
print(f"'   '.isspace():    {'   '.isspace()}")       # True
print(f"'Hello'.istitle():  {'Hello'.istitle()}")     # True

# Practical use: input validation
user_input = "42"
if user_input.isdigit():
    print(f"\n{user_input} is a valid number")
````

## 📝 Key Takeaways: Java → Python

| Concept | Java | Python |
|---------|------|--------|
| Immutable? | Yes | Yes |
| Char type | `char c = 'A';` | No char — `s[0]` is still a `str` |
| Upper/Lower | `toUpperCase()` | `upper()` / `lower()` |
| Trim | `trim()` / `strip()` (Java 11) | `strip()` |
| Contains | `str.contains("x")` | `"x" in str` |
| Split | `str.split(",")` | `str.split(",")` |
| Join | `String.join("-", list)` | `"-".join(list)` ⚠️ reversed! |
| Substring | `str.substring(0, 5)` | `str[0:5]` (slicing) |
| Reverse | `StringBuilder.reverse()` | `str[::-1]` |
| Format | `String.format("%s", x)` | `f"{x}"` (f-strings) |
| Multi-line | Text blocks (Java 15+) | `"""triple quotes"""` |
| Char checks | `Character.isDigit(c)` | `str.isdigit()` (whole string) |
