# 7. Exception Handling

Python's exception handling is similar to Java's `try-catch-finally`, but with some important differences: Python has **no checked exceptions** (no `throws` clause!), uses `except` instead of `catch`, and has a unique `else` clause.

This notebook covers: try/except/else/finally, raising exceptions, common exception types, catching multiple exceptions, and the key differences from Java.

### 7.1 Basic try/except

**☕ JAVA:**
```java
try {
    int result = 10 / 0;
} catch (ArithmeticException e) {
    System.out.println("Error: " + e.getMessage());
}
```

**🐍 PYTHON:** `catch` → `except`, access exception with `as`.
```python
try:
    result = 10 / 0
except ZeroDivisionError as e:
    print(f"Error: {e}")
```

````python
# Basic try/except
try:
    result = 10 / 0
except ZeroDivisionError:
    print("Cannot divide by zero!")

# Accessing the exception object
try:
    numbers = [1, 2, 3]
    print(numbers[10])
except IndexError as e:
    print(f"Index error: {e}")
````

### 7.2 The Full try/except/else/finally

**☕ JAVA:** Has `try-catch-finally` but **no `else` clause**.

**🐍 PYTHON:** Adds `else` — runs only if **no exception** was raised.

```python
try:
    ...           # Code that might fail
except SomeError:
    ...           # Handle the error
else:
    ...           # Runs ONLY if no exception occurred ← Python-only!
finally:
    ...           # Always runs (cleanup)
```

````python
def divide(a: float, b: float) -> float:
    try:
        result = a / b
    except ZeroDivisionError:
        print("  Cannot divide by zero!")
        return 0.0
    except TypeError as e:
        print(f"  Type error: {e}")
        return 0.0
    else:
        # Only runs if NO exception occurred
        print(f"  Division successful: {result}")
        return result
    finally:
        # Always runs, even after return!
        print("  Cleanup (always runs)")

print("--- Success case ---")
divide(10, 2)

print("\n--- Error case ---")
divide(10, 0)
````

### 7.3 Raising Exceptions

**☕ JAVA:**
```java
throw new IllegalArgumentException("Age cannot be negative: " + age);
```

**🐍 PYTHON:** `throw` → `raise`, no `new` keyword needed.
```python
raise ValueError(f"Age cannot be negative: {age}")
```

````python
def validate_age(age: int) -> None:
    if age < 0:
        raise ValueError(f"Age cannot be negative: {age}")
    if age > 150:
        raise ValueError(f"Age seems unrealistic: {age}")
    print(f"Age {age} is valid")

# Test it
validate_age(25)     # OK

try:
    validate_age(-5)  # Raises ValueError
except ValueError as e:
    print(f"Caught: {e}")
````

````python
# Re-raising exceptions — handle partially, then pass it on
def process_data(data):
    try:
        result = int(data)
    except ValueError:
        print(f"  Logging: invalid data '{data}'")
        raise  # Re-raise the same exception

try:
    process_data("not_a_number")
except ValueError as e:
    print(f"Caught re-raised: {e}")
````

### 7.4 Catching Multiple Exceptions

**☕ JAVA 7+:** `catch (IOException | SQLException e) { ... }`

**🐍 PYTHON:** Use a tuple: `except (ValueError, TypeError):`

````python
def safe_convert(value):
    """Try to convert value to int."""
    try:
        return int(value)
    except (ValueError, TypeError) as e:
        print(f"Conversion failed: {e}")
        return None

print(safe_convert("42"))       # 42
print(safe_convert("hello"))    # None (ValueError)
print(safe_convert(None))       # None (TypeError)
````

### 7.5 Common Built-in Exceptions

Python has many built-in exception types. Here are the most common ones:

| Exception | When it happens | Java equivalent |
|-----------|----------------|------------------|
| `ValueError` | Wrong value (e.g., `int("abc")`) | `IllegalArgumentException` |
| `TypeError` | Wrong type (e.g., `"a" + 1`) | `ClassCastException` |
| `KeyError` | Missing dict key | N/A (returns null) |
| `IndexError` | List index out of range | `IndexOutOfBoundsException` |
| `FileNotFoundError` | File doesn't exist | `FileNotFoundException` |
| `AttributeError` | Object has no attribute | N/A (compile error in Java) |
| `ZeroDivisionError` | Division by zero | `ArithmeticException` |
| `NameError` | Undefined variable | N/A (compile error in Java) |

````python
# Quick demo of common exceptions

# KeyError — missing dict key
person = {"name": "Alice"}
try:
    print(person["age"])
except KeyError as e:
    print(f"KeyError: key {e} not found")
    print(f"  Tip: use .get() instead: {person.get('age', 'N/A')}")

# AttributeError — object has no attribute
try:
    "hello".append("!")
except AttributeError as e:
    print(f"\nAttributeError: {e}")
````

### 7.6 Catch-All with `Exception`

**☕ JAVA:** `catch (Exception e)` — catches all exceptions.

**🐍 PYTHON:** `except Exception as e:` — same idea.

> ⚠️ **Use sparingly!** Catching `Exception` can hide bugs. Prefer catching specific exceptions when possible.

````python
# Catch-all — useful for logging/graceful degradation
def safe_operation(func, *args):
    try:
        return func(*args)
    except Exception as e:
        print(f"Error ({type(e).__name__}): {e}")
        return None

# Test with different failures
safe_operation(int, "abc")         # ValueError
safe_operation(lambda: 1/0)        # ZeroDivisionError
safe_operation(lambda: [][5])      # IndexError
````

### 7.7 No Checked Exceptions! 🎉

One of the **biggest differences** from Java:

**☕ JAVA:** Must declare checked exceptions or handle them:
```java
public void readFile(String path) throws IOException {
    // Compiler forces you to handle or declare
}
```

**🐍 PYTHON:** No such thing! All exceptions are unchecked:
```python
def read_file(path):
    # No throws clause needed — ever!
    with open(path) as f:
        return f.read()
```

This means:
- ✅ Less boilerplate — no `throws` declarations everywhere
- ⚠️ More responsibility — you need to **know** which exceptions a function can raise (read the docs!)

## 📝 Key Takeaways: Java → Python

| Concept | Java | Python |
|---------|------|--------|
| Catch | `catch (Exception e)` | `except Exception as e:` |
| Throw | `throw new Exception()` | `raise Exception()` |
| Re-throw | `throw;` (in catch) | `raise` (in except) |
| Finally | `finally { ... }` | `finally:` |
| Else (no error) | ❌ Not available | `else:` ← Python-only! |
| Checked exceptions | Yes (`throws IOException`) | ❌ None — all unchecked |
| Multi-catch | `catch (A \| B e)` | `except (A, B) as e:` |
| Wrong value | `IllegalArgumentException` | `ValueError` |
| Wrong type | `ClassCastException` | `TypeError` |
| Missing key | Returns `null` | `KeyError` (or use `.get()`) |
| Index OOB | `IndexOutOfBoundsException` | `IndexError` |
