# 5. Functions

Python functions are simpler than Java methods — no access modifiers, no class required, no return type declaration (though type hints are optional). Functions are also **first-class objects**, meaning they can be stored in variables and passed as arguments.

This notebook covers: defining functions, default/keyword arguments, `*args`/`**kwargs`, lambda, multiple return values, first-class functions, and docstrings.

### 5.1 Basic Functions

**☕ JAVA:**

```java
public int add(int a, int b) {
    return a + b;
}
```

**🐍 PYTHON:**

```python
def add(a, b):
    return a + b

# Or with type hints:
def add(a: int, b: int) -> int:
    return a + b
```

````python
def greet(name: str) -> str:
    """Return a greeting message."""
    return f"Hello, {name}!"

print(greet("Alice"))
print(greet("Bob"))
````

### 5.2 Default Parameters

**☕ JAVA:** No default parameters — you need method overloading:
```java
public String greet(String name) { return greet(name, "Hello"); }
public String greet(String name, String greeting) { return greeting + ", " + name; }
```

**🐍 PYTHON:** Default values built into the signature.
```python
def greet(name, greeting="Hello"):
    return f"{greeting}, {name}!"
```

````python
def greet(name: str, greeting: str = "Hello") -> str:
    """Return a greeting message with optional custom greeting."""
    return f"{greeting}, {name}!"

print(greet("Alice"))                       # Uses default: Hello
print(greet("Bob", "Hi"))                    # Positional override
print(greet(greeting="Hey", name="Charlie")) # Named (keyword) arguments — any order!
````

### 5.3 `*args` — Variable Positional Arguments

**☕ JAVA (varargs):**
```java
public void log(String... messages) {
    for (String msg : messages) { ... }
}
```

**🐍 PYTHON:** `*args` collects extra positional arguments into a **tuple**.
```python
def log(*messages):
    for msg in messages: ...
```

````python
def log(*messages, level="INFO"):
    """Accept any number of messages."""
    for msg in messages:
        print(f"  [{level}] {msg}")

log("Starting", "Processing", "Done")
print()
log("Error occurred", level="ERROR")
````

### 5.4 `**kwargs` — Variable Keyword Arguments

**☕ JAVA:** No direct equivalent — you'd use a `Map<String, Object>` parameter.

**🐍 PYTHON:** `**kwargs` collects extra keyword arguments into a **dict**.
```python
def create_user(**kwargs):
    return kwargs  # {'name': 'Alice', 'age': 30}
```

````python
def create_user(**kwargs):
    """Accept any keyword arguments as a dict."""
    return kwargs

user = create_user(name="Alice", age=30, email="alice@example.com")
print(f"User: {user}")
print(f"Name: {user['name']}")
````

### 5.5 Multiple Return Values

**☕ JAVA (before Java 16):** To return multiple values, you had to create an entire class with boilerplate:
```java
public final class MinMax {
    private final int min;
    private final int max;

    public MinMax(int min, int max) {
        this.min = min;
        this.max = max;
    }

    public int min() { return min; }
    public int max() { return max; }

    // Plus equals(), hashCode(), toString()...
}
```

**☕ JAVA 16+ (record):** Much cleaner — one line replaces all of the above:
```java
public record MinMax(int min, int max) {}
```
The compiler auto-generates: `private final` fields, constructor, getters (accessors), `equals()`, `hashCode()`, and `toString()`.

**🐍 PYTHON:** Just return multiple values — Python packs them into a tuple automatically. No class needed!
```python
def find_min_max(nums):
    return min(nums), max(nums)
```

````python
def find_min_max(numbers):
    """Return both the min and max of a list."""
    return min(numbers), max(numbers)

# Unpack the results
lo, hi = find_min_max([4, 1, 7, 2, 9, 3])
print(f"Min: {lo}, Max: {hi}")

# Or keep as tuple
result = find_min_max([4, 1, 7, 2, 9, 3])
print(f"Result tuple: {result}")
````

### 5.6 Lambda Functions (Anonymous Functions)

**☕ JAVA:**
```java
Function<Integer, Integer> square = n -> n * n;
list.sort((a, b) -> a.compareTo(b));
```

**🐍 PYTHON:** `lambda` — a one-line anonymous function.
```python
square = lambda n: n * n
list.sort(key=lambda x: x.name)
```

Lambdas are limited to a **single expression** — for anything complex, use `def`.

````python
# Lambda basics
square = lambda n: n * n
print(f"square(5) = {square(5)}")

add = lambda a, b: a + b
print(f"add(3, 4) = {add(3, 4)}")

# Most common use: as a key function for sorting
names = ["Charlie", "Alice", "Bob"]
names.sort(key=lambda name: len(name))
print(f"\nSorted by length: {names}")

# Sort a list of tuples by second element
students = [("Alice", 85), ("Bob", 92), ("Charlie", 78)]
students.sort(key=lambda s: s[1])
print(f"Sorted by grade: {students}")
````

### 5.7 Functions as First-Class Objects

In Python, functions are **objects** — you can store them in variables, pass them as arguments, and return them from other functions.

**☕ JAVA:** Requires functional interfaces (`Function<T, R>`, `Predicate<T>`, etc.).

**🐍 PYTHON:** Functions are just values — no special syntax needed.

````python
# Store function in a variable
def shout(text):
    return text.upper() + "!!!"

def whisper(text):
    return text.lower() + "..."

# Choose function at runtime
speak = shout
print(speak("hello"))         # HELLO!!!

speak = whisper
print(speak("hello"))         # hello...
````

````python
# Pass function as argument
def apply_to_list(func, items):
    """Apply a function to each item and return results."""
    return [func(item) for item in items]

numbers = [1, 2, 3, 4, 5]
print(f"Doubled: {apply_to_list(lambda x: x * 2, numbers)}")
print(f"Squared: {apply_to_list(lambda x: x ** 2, numbers)}")
````

### 5.8 Docstrings (Python's Javadoc)

**☕ JAVA:** Javadoc comments `/** ... */` above methods.

**🐍 PYTHON:** Triple-quoted strings as the **first line** inside a function. Accessible at runtime via `__doc__`.

````python
def calculate_bmi(weight_kg: float, height_m: float) -> float:
    """
    Calculate Body Mass Index (BMI).
    
    Args:
        weight_kg: Weight in kilograms
        height_m: Height in meters
    
    Returns:
        BMI as a float
    """
    return weight_kg / (height_m ** 2)

print(f"BMI: {calculate_bmi(70, 1.75):.1f}")

# Access docstring at runtime
print(f"\nDocstring:\n{calculate_bmi.__doc__}")

# Also works with help()
help(calculate_bmi)
````

## 📝 Key Takeaways: Java → Python

| Concept | Java | Python |
|---------|------|--------|
| Define function | `public int add(int a, int b)` | `def add(a, b):` |
| Type hints | Required | Optional: `def add(a: int) -> int:` |
| Default params | Method overloading | `def f(x, y=10):` |
| Named args | N/A | `f(name="Alice", age=30)` |
| Varargs | `String... args` | `*args` (tuple) |
| Keyword varargs | `Map<String, Object>` | `**kwargs` (dict) |
| Multiple returns | Record / array | `return a, b` (tuple) |
| Lambda | `n -> n * n` | `lambda n: n * n` |
| First-class func | `Function<T,R>` interface | Functions are just values |
| Documentation | `/** Javadoc */` | `"""docstring"""` |
| Access docs | N/A (compile-time only) | `func.__doc__` / `help(func)` |
