# 2. Control Flow

Python's control flow is cleaner than Java's — no parentheses around conditions, no braces, just colons and indentation.

This notebook covers: if/elif/else, ternary expressions, for loops, while loops, break/continue, and the unique `for...else` pattern.

### 2.1 If / Elif / Else

**☕ JAVA:**
```java
if (x > 0) {
    // ...
} else if (x < 0) {
    // ...
} else {
    // ...
}
```

**🐍 PYTHON:** No parentheses, use colon `:`, indentation defines the block.
```python
if x > 0:
    ...
elif x < 0:
    ...
else:
    ...
```

````python
x = 10
if x > 0:
    print("Positive")
elif x < 0:
    print("Negative")
else:
    print("Zero")
````

### 2.2 Ternary (Conditional Expression)

**☕ JAVA:**
```java
String status = age >= 18 ? "adult" : "minor";
```

**🐍 PYTHON:**
```python
status = "adult" if age >= 18 else "minor"
```

Notice: Python reads like English — *"adult if age >= 18, else minor"*.

````python
age = 25
status = "adult" if age >= 18 else "minor"
print(f"Age {age} → {status}")

# Another example
score = 85
result = "pass" if score >= 60 else "fail"
print(f"Score {score} → {result}")
````

### 2.3 For Loops

**☕ JAVA:**
```java
for (int i = 0; i < 5; i++) { ... }
for (String name : names) { ... }
```

**🐍 PYTHON:**
```python
for i in range(5): ...      # range-based
for name in names: ...      # for-each (most common)
```

````python
# Range-based loop
print("range(5):")
for i in range(5):          # 0, 1, 2, 3, 4
    print(f"  i = {i}")
````

````python
# range(start, stop, step)
print("range(2, 10, 3):")
for i in range(2, 10, 3):   # 2, 5, 8
    print(f"  i = {i}")

# Counting down
print("\nrange(5, 0, -1):")
for i in range(5, 0, -1):   # 5, 4, 3, 2, 1
    print(f"  i = {i}")
````

````python
# For-each — the most common pattern
for name in ["Alice", "Bob", "Charlie"]:
    print(f"  Hello, {name}")
````

### 2.4 enumerate() and zip()

**☕ JAVA:**
```java
for (int i = 0; i < names.size(); i++) {
    System.out.println(i + ": " + names.get(i));
}
```

**🐍 PYTHON:** Use `enumerate()` — no manual index tracking!
```python
for i, name in enumerate(names):
    print(f"{i}: {name}")
```

````python
# enumerate() — loop with index
fruits = ["apple", "banana", "cherry"]
for i, fruit in enumerate(fruits):
    print(f"  {i}: {fruit}")

# Start index from 1
print()
for i, fruit in enumerate(fruits, start=1):
    print(f"  {i}. {fruit}")
````

````python
# zip() — loop over multiple sequences in parallel
names = ["Alice", "Bob", "Charlie"]
ages = [30, 25, 35]
cities = ["NYC", "LA", "Chicago"]

for name, age, city in zip(names, ages, cities):
    print(f"  {name} is {age}, lives in {city}")
````

### 2.5 While Loops

**☕ JAVA:**
```java
while (count > 0) {
    count--;
}
```

**🐍 PYTHON:**
```python
while count > 0:
    count -= 1  # No -- operator in Python!
```

````python
count = 3
while count > 0:
    print(f"  Countdown: {count}")
    count -= 1  # Python has no ++ or -- operators
print("Go!")
````

### 2.6 break, continue, and pass

These work the same as Java:
- `break` — exit the loop immediately
- `continue` — skip to the next iteration
- `pass` — do nothing (placeholder for empty blocks) — **unique to Python**

````python
# break — stop when we find what we want
print("break example:")
for n in range(1, 20):
    if n == 5:
        print(f"  Found {n}, stopping!")
        break
    print(f"  checking {n}...")
````

````python
# continue — skip odd numbers
print("continue example (even numbers only):")
for n in range(10):
    if n % 2 != 0:
        continue
    print(f"  {n}")
````

````python
# pass — placeholder for empty blocks
# In Java you can have empty braces: if (x > 0) { }
# In Python, an empty block is a syntax error — you need 'pass'

x = 10
if x > 0:
    pass  # TODO: implement this later

# Without 'pass', this would fail:
# if x > 0:
#     <-- IndentationError: expected an indented block

print("pass is a no-op — program continues normally")
````

### 2.7 for...else / while...else (Unique to Python!)

**☕ JAVA:** No equivalent — you'd need a boolean flag.

**🐍 PYTHON:** The `else` block runs only if the loop **completes without hitting `break`**.

````python
# for...else — else runs if NO break occurred
target = 7
for n in range(1, 6):    # 1 through 5 — won't find 7
    if n == target:
        print(f"Found {target}!")
        break
else:
    print(f"{target} was NOT found in the range")

print()

# Now try with a target that exists
target = 3
for n in range(1, 6):
    if n == target:
        print(f"Found {target}!")
        break
else:
    print(f"{target} was NOT found in the range")
````

## 📝 Key Takeaways: Java → Python

| Concept | Java | Python |
|---------|------|--------|
| If statement | `if (x > 0) { }` | `if x > 0:` |
| Else if | `else if` | `elif` |
| Ternary | `cond ? a : b` | `a if cond else b` |
| For loop | `for (int i = 0; i < n; i++)` | `for i in range(n):` |
| For-each | `for (String s : list)` | `for s in list:` |
| Loop with index | Manual `i++` | `enumerate()` |
| Parallel iteration | N/A (manual) | `zip()` |
| Decrement | `count--` | `count -= 1` |
| Empty block | `{ }` | `pass` |
| Loop completed? | Boolean flag | `for...else` |
