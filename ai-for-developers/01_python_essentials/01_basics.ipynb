## 1. PYTHON BASICS - THE FUNDAMENTALS

### 1.1 No Main Method, No Semicolons, No Braces

**☕ JAVA:**
```java
public class Main {
    public static void main(String[] args) {
        System.out.println("Hello");
    }
}
```

**🐍 PYTHON:**
```python
print("Hello")
```

That's it. Python uses **indentation** instead of braces, and there's no `main` method or class wrapper needed.
```

````python
# Your first Python line — run this cell!
print("Hello from Python!")
````

### 1.2 Variables — No Type Declarations Needed

**☕ JAVA:**
```java
String name = "Alice";
int age = 30;
double price = 19.99;
boolean isActive = true;
```

**🐍 PYTHON:**
```python
name = "Alice"
age = 30
price = 19.99
is_active = True  # Note: True/False are capitalized
```

````python
# Variables — no type declarations needed
name = "Alice"              # Java: String name = "Alice";
age = 30                    # Java: int age = 30;
price = 19.99               # Java: double price = 19.99;
is_active = True            # Java: boolean isActive = true;

print(f"Name: {name}, Age: {age}, Active: {is_active}")
````

````python
# Type hints are optional — they don't change runtime behavior,
# but help editors and tools catch bugs (like TypeScript annotations)
name: str = "Alice"
age: int = 30
prices: list[float] = [9.99, 19.99, 29.99]

print(f"name is still: {name}")
print(f"prices: {prices}")
````

### 1.3 Dynamic Typing

**☕ JAVA:** Variables are locked to their declared type.

**🐍 PYTHON:** Variables can be reassigned to **any type** at any time — the type lives on the *value*, not the variable.

````python
# Dynamic typing — variables can change type!
x = 42
print(f"x = {x}, type: {type(x)}")   # <class 'int'>

x = "hello"
print(f"x = {x}, type: {type(x)}")   # <class 'str'>

x = [1, 2, 3]
print(f"x = {x}, type: {type(x)}")   # <class 'list'>

# Java would throw a compile error for this!
````

### 1.4 Multiple Assignment & Swapping

**☕ JAVA:**
```java
int a = 1, b = 2, c = 3;
// Swap requires a temp variable
int temp = a; a = b; b = temp;
```

**🐍 PYTHON:**
```python
a, b, c = 1, 2, 3
a, b = b, a  # Swap — no temp needed!
```

````python
# Multiple assignment (tuple unpacking)
a, b, c = 1, 2, 3
print(f"a={a}, b={b}, c={c}")

# Swap without temp variable!
a, b = b, a
print(f"After swap: a={a}, b={b}")

# Same value to multiple variables
x = y = z = 0
print(f"x={x}, y={y}, z={z}")
````

### 1.5 Print and F-Strings

**☕ JAVA:**
```java
System.out.println("Name: " + name + ", Age: " + age);
System.out.printf("Name: %s, Age: %d%n", name, age);
```

**🐍 PYTHON:**
```python
print(f"Name: {name}, Age: {age}")  # F-strings (Python 3.6+)
```

````python
# F-strings are powerful
name = "Alice"  # Redefine here so this cell works independently
age = 30
price = 19.99

value = 3.14159
print(f"Pi = {value:.2f}")              # Format to 2 decimals
print(f"{'centered':^20}")              # Center in 20 chars
print(f"{name.upper() = }")             # Debug format (shows expression)

# Multi-line f-strings
message = (
    f"User: {name}\n"
    f"Age:  {age}\n"
    f"Price: ${price:.2f}"
)
print(message)
````

### 1.6 Number Types

**☕ JAVA:** `int`, `long`, `float`, `double` — all fixed precision.

**🐍 PYTHON:** `int` has **arbitrary precision** (no overflow!), `float` is 64-bit (like Java's `double`).

````python
# Python int has NO overflow — try this in Java!
big = 2 ** 100
print(f"2^100 = {big}")
print(f"Digits: {len(str(big))}")

# Float (64-bit, like Java's double)
pi = 3.14159
print(f"Pi: {pi}")

# Underscores for readability (like Java)
million = 1_000_000
print(f"Million: {million}")

# Division differences
print(f"7 / 2  = {7 / 2}")    # 3.5  — always float!
print(f"7 // 2 = {7 // 2}")   # 3    — floor division (like Java's int division)
print(f"7 % 2  = {7 % 2}")    # 1    — modulo
print(f"2 ** 8 = {2 ** 8}")   # 256  — power (Java uses Math.pow())
````

### 1.7 Operators

| Java | Python | Meaning |
|------|--------|--------|
| `&&` | `and` | Logical AND |
| `\|\|` | `or` | Logical OR |
| `!` | `not` | Logical NOT |
| `Math.pow(a, b)` | `a ** b` | Power |
| `a / b` (int) | `a // b` | Floor division |
| `==` | `==` | Value equality |
| `==` (objects) | `is` | Identity (same object) |

````python
# Logical operators — words, not symbols!
x = 10
print(f"x > 5 and x < 20: {x > 5 and x < 20}")   # Java: x > 5 && x < 20
print(f"not (x > 15):     {not (x > 15)}")         # Java: !(x > 15)
print(f"x < 5 or x > 8:  {x < 5 or x > 8}")       # Java: x < 5 || x > 8

# Chained comparisons — unique to Python!
age = 25
print(f"18 <= age <= 65: {18 <= age <= 65}")  # Java: age >= 18 && age <= 65

# Identity vs Equality
a = [1, 2, 3]
b = [1, 2, 3]
print(f"a == b: {a == b}")  # True  — same value
print(f"a is b: {a is b}")  # False — different objects
````

### 1.8 None vs null

**☕ JAVA:**
```java
Object result = null;
if (result == null) { ... }
```

**🐍 PYTHON:**
```python
result = None
if result is None: ...  # Use 'is', not '=='
```

````python
result = None  # Java: Object result = null;
if result is None:
    print("Result is None")

# Use 'is' for None comparisons, not ==
# if result is None:  ✓  (checks identity)
# if result == None:  ✗  (works but not Pythonic)
````

### 1.9 Boolean & Truthy/Falsy Values

**☕ JAVA:** Only `boolean` can be used in conditions.

**🐍 PYTHON:** **Any object** can be used in a boolean context. These are **falsy** (evaluate to `False`):
- `None`, `False`
- `0`, `0.0`, `0j`
- `""` (empty string), `[]`, `()`, `{}`, `set()`

````python
# Everything is truthy except the falsy values listed above
print(f"  bool(0)       = {bool(0)}")         # False
print(f"  bool(1)       = {bool(1)}")         # True
print(f"  bool('')      = {bool('')}")        # False
print(f"  bool('hello') = {bool('hello')}")   # True
print(f"  bool([])      = {bool([])}")        # False
print(f"  bool([1, 2])  = {bool([1, 2])}")   # True
print(f"  bool(None)    = {bool(None)}")      # False

# Practical usage — no need for explicit length checks!
items = []
if not items:  # Pythonic! (instead of: if len(items) == 0)
    print("\nList is empty")

name = ""
if not name:   # Pythonic! (instead of: if name == "")
    print("Name is empty")
````

### 1.10 User Input

**☕ JAVA:**
```java
Scanner sc = new Scanner(System.in);
String name = sc.nextLine();
int age = sc.nextInt();
```

**🐍 PYTHON:**
```python
name = input("Enter name: ")
age = int(input("Enter age: "))  # input() always returns str!
```

````python
# Try it — input() works interactively in Jupyter!
name = input("Enter your name: ")
age = int(input("Enter your age: "))  # input() returns str, so cast to int
print(f"Hello {name}, you will be {age + 10} in 10 years!")
````

````python
# Why int() is needed: input() ALWAYS returns a string
age_str = "30"             # This is what input() gives you
print(f"type of age_str: {type(age_str)}")  # <class 'str'>

age = int(age_str)          # Must convert explicitly
print(f"type of age:     {type(age)}")      # <class 'int'>
print(f"age + 10 = {age + 10}")             # Now math works!
````

## 📝 Key Takeaways: Java → Python

| Concept | Java | Python |
|---------|------|--------|
| Entry point | `public static void main(String[] args)` | Just write code |
| Blocks | `{ }` | Indentation (4 spaces) |
| Statements | End with `;` | No semicolons |
| Variables | `int x = 5;` | `x = 5` |
| String formatting | `String.format()` / `+` | `f"...{var}"` |
| Null | `null` | `None` (check with `is`) |
| Booleans | `true` / `false` | `True` / `False` |
| Logical ops | `&&` `\|\|` `!` | `and` `or` `not` |
| Integer overflow | Yes (wraps at 2³¹) | Never (arbitrary precision) |
| Floor division | `7 / 2 → 3` | `7 // 2 → 3` |
| Power | `Math.pow(2, 8)` | `2 ** 8` |
| Swap | Needs temp variable | `a, b = b, a` |
| User input | `Scanner` | `input()` (returns `str`) |
| Falsy values | Only `false` | `0`, `""`, `[]`, `None`, etc. |
